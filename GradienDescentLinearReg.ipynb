```python
import pandas as pd
import matplotlib.pyplot as plt

data ={
    'x' : [20,40,60,80,100,120,140,160,180,200,220,240,260,280,300,320,340,360,380,400],
    'y' : [10,25,15,40,32,50,39,65,55,75,60,85,78,95,82,105,98,115,100,125]
    
}

df = pd.DataFrame(data)
df

```

```python
x = df[['x']]
y = df[['y']]
```

```python
plt.plot(x,y)
plt.scatter(x,y,color='red')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
df['y_pred'] = reg.predict(x)
df
```

```python
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x,y,color='green',label='Actual Plot')
plt.scatter(x,df['y_pred'],color='red' ,label='Predict without G.D')
plt.plot(x, df['y_pred'],color='black', label='Predicted Line without G.D')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
#Gradient Descent Algorithm apply to minimum error


x = [20, 40, 60, 80, 100, 120, 140, 160, 180, 200,
         220, 240, 260, 280, 300, 320, 340, 360, 380, 400]
y = [10, 25, 15, 40, 32, 50, 39, 65, 55, 75,
     60, 85, 78, 95, 82, 105, 98, 115, 100, 125]

x_min = min(x)
x_max = max(x)
x = [(xi - x_min) / (x_max - x_min) for xi in x]

#  Gradient Descent
m = 0
c = 0
learning_rate = 0.01
epochs = 1000

for i in range(epochs):
    total_error_m = 0
    total_error_c = 0
    n = len(x)

    for j in range(n):
        y_pred = m * x[j] + c
        error = y_pred - y[j]
        total_error_m += error * x[j]
        total_error_c += error

    dm = total_error_m / n
    dc = total_error_c / n

    m = m -  learning_rate * dm
    c= c - learning_rate * dc

# Final formula
print(f"Final formula: y = {m:.2f} * x + {c:.2f}")

# Predict for speed = 400
x_input = 400
x_norm = (x_input - x_min) / (x_max - x_min)
y_pred_input = m * x_norm + c
print(f"Predicted risk for speed = {x_input} is: {y_pred_input:.2f}")


```

```python
#Plot Actual vs Predicted
y_pred_all = [m * xi + c for xi in x]

plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x,y,color='green',label='Actual Plot')

plt.scatter(x,df['y_pred'],color='red' ,label='Predict without G.D')
plt.plot(x, df['y_pred'],color='black', label='Predicted Line without G.D')

plt.scatter(x,y_pred_all,color='black', label='Predict with G.D')
plt.plot(x, y_pred_all, color='red', label='Predicted Line with G.D')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Actual vs Predicted")
plt.legend()
plt.grid()
plt.show()
```
