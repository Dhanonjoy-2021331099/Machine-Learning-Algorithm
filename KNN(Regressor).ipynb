```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Dhanonjoy-2021331099/Machine-Learning-Algorithm/refs/heads/main/KNN(Regression).csv")
df
```

```python
# Find Age = 38, Height=5.5 , then Weight = ?
```

```python
new_point = [38,5.5]
import numpy as np
def distance(row):
    return np.sqrt((row['Age'] - new_point[0])**2 + (row['Height']-new_point[1])**2)
df['Distance'] = df.apply(distance,axis=1)
df
```

```python
rank = df.sort_values(by="Distance")
df['Rank']= df['Distance'].rank(method='dense').astype(int)
df
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = df[['Age', 'Height']]  # Input features
y = df['Weight']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
y_pred= knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
print("R2 Score:", r2_score(y_test, y_pred))
```
